# ### Creating df for holding the data of historic averages

```python
import pandas as pd
import calendar

# Check if the cyclone exists in the onset_df
if cyclone_name in onset_df['Cyclone'].values:
    # Initialize an empty DataFrame to store the average values
    hist_avg_df = pd.DataFrame()

    # Get the onset date for the cyclone
    cyclone_onset = pd.to_datetime(onset_df.loc[onset_df['Cyclone'] == cyclone_name, 'Onset_Date'].values[0])

    # Calculate averages for each date from 2013 to 2021
    for year in range(2013, 2022):
        # Loop through the desired months and dates
        for month in range(1, 13):
            # Get the number of days in the month
            num_days = calendar.monthrange(year, month)[1]

            for day in range(1, num_days + 1):
                # Construct the date string
                current_date = pd.to_datetime(f"{year}-{month:02d}-{day:02d}")

                # Select the data for the specific 'settl_pcod'
                if grid3 in df_date.index:
                    specific_data = df_date.loc[grid3, current_date.strftime('%Y-%m-%d')]

                    # Store the average value in the DataFrame
                    hist_avg_df.at[f'{month:02d}-{day:02d}', year] = specific_data.mean()

    # Calculate the overall average for each date across the years
    hist_avg_df[f'{cyclone_name} Historical Average'] = hist_avg_df.mean(axis=1)

    # Display the DataFrame
    print(hist_avg_df)

else:
    print(f"Cyclone name '{cyclone_name}' not found in the dataset.")
```

```python
# Assuming 'cyclone_name', 'onset_df', 'hist_avg_df', 'df_date', 'grid3', 'lag' are already defined

# Check if the cyclone exists in the onset_df
if cyclone_name in onset_df['Cyclone'].values:
    # Get the onset date for the cyclone
    cyclone_onset = pd.to_datetime(onset_df.loc[onset_df['Cyclone'] == cyclone_name, 'Onset_Date'].values[0])

    # Calculate the start and end dates based on the lag
    start_date = cyclone_onset - pd.Timedelta(days=lag)
    end_date = cyclone_onset + pd.Timedelta(days=lag)
    date_range = pd.date_range(start=start_date, end=end_date)

    # Filter the hist_avg_df for the selected date range
    avg_data = hist_avg_df.loc[hist_avg_df.index.isin(date_range.strftime('%m-%d'))]

    # Convert the index to only month-day format for plotting
    avg_data.index = [date.strftime('%m-%d') for date in date_range]

    # Split the data into pre and post cyclone onset
    cyclone_onset_str = cyclone_onset.strftime('%m-%d')
    pre_data = avg_data.loc[avg_data.index < cyclone_onset_str]
    post_data = avg_data.loc[avg_data.index >= cyclone_onset_str]
    X_pre = np.arange(len(pre_data)).reshape(-1, 1)
    y_pre = pre_data[f'{cyclone_name} Historical Average'].values.reshape(-1, 1)
    X_post = np.arange(len(post_data)).reshape(-1, 1)
    y_post = post_data[f'{cyclone_name} Historical Average'].values.reshape(-1, 1)

    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(avg_data.index, avg_data[f'{cyclone_name} Historical Average'], label='Historical Average')

    # Fit and plot linear regression models if data is available
    if len(pre_data) > 0:
        model_pre = LinearRegression()
        model_pre.fit(X_pre, y_pre)
        plt.plot(pre_data.index, model_pre.predict(X_pre), color='blue', linestyle='--', label='Hist Pre Regression Line')

    if len(post_data) > 0:
        model_post = LinearRegression()
        model_post.fit(X_post, y_post)
        plt.plot(post_data.index, model_post.predict(X_post), color='green', linestyle='--', label='Hist Post Regression Line')

    plt.title(f'{cyclone_name} - Historical Average Data Analysis')
    plt.xlabel('Date (Month-Day)')
    plt.ylabel('Average Values')
    plt.xticks(rotation=45)
    plt.axvline(x=cyclone_onset_str, color='red', linestyle='-', label=f'{cyclone_name} Onset Date')
    plt.legend()
    plt.show()

else:
    print(f"Cyclone name '{cyclone_name}' not found in the dataset.")
```
